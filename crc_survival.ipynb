```python
import pandas as pd
```

```python
crc_data = pd.read_excel("C:/Users\dnjsw/python/dataset/191201_dataset_28feature.xlsx")
```

```python
# 데이터 결측치 제거
crc_data_dropna = crc_data.dropna()
crc_data_dropna.info()
```

```python
# X, y 나누기
X_crc = crc_data_dropna[["Age","Sex","ASA","BMI","DM_history","Pulmonary_disease",
                        "Liver_disease","Heart_disease","Kidney_disease",
                       "Prior_Dx_cancer","Initial_CEA","Perforation",
                       "Obstruction","Emergency",
                       "Histologic_type","pT","LVI","Distal resection margin",
                        "pN","Harvested_LN","Positive_LN",
                       "pM","pTNM","BRAF","Early_Complication","Postop_Chemotherapy",
                       "Postop_Chemo_Regimen"]]
y_crc = crc_data_dropna[["OS"]]
```

```python
# data type 나누기
category = X_crc[["Sex","ASA","DM_history","Pulmonary_disease",
                  "Liver_disease","Heart_disease","Kidney_disease",
                  "Prior_Dx_cancer","Perforation",
                  "Obstruction","Emergency",
                  "Histologic_type","pT","LVI","Distal resection margin",
                  "pN","pM","pTNM","BRAF","Early_Complication","Postop_Chemotherapy",
                  "Postop_Chemo_Regimen"]]
numeric = X_crc[["Age","BMI","Initial_CEA","Harvested_LN","Positive_LN"]]
```

```python
# category 데이터 타입 변경
category = category.astype(object)
```

```python
# category one-hot encoding
category = pd.get_dummies(category)
category
```

```python
# numeric data minmax normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
numeric[:] = scaler.fit_transform(numeric[:])
```

```python
numeric
```

```python
#전처리된 데이터 합치기
X_crc = pd.concat([category, numeric],axis=1)
X_crc
```

```python
from sklearn.model_selection import train_test_split
```

```python
Train_X, Test_X, Train_y, Test_y = train_test_split(X_crc,y_crc)
```
